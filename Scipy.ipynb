(x^{2}+y^{2}-1)^{3}-x^{2}y^{3}=0

```python
def lemniscate(x, y):
    return (x**2 + y**2 - 1)**3 - x**2 * y**3
```

```python
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 300)
x, y = np.meshgrid(x, y)
z = lemniscate(x, y)
```

```python
plt.contour(x, y, z, levels=[0], colors='r')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.title('Lemniscate of Bernoulli')
plt.grid(True)
plt.show()
```
